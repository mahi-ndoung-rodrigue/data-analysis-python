# Task 3 - GroupBy and Pivot Tables

Group the parameters "GPU", "CPU_core" and "Price" to make a pivot table and visualize this connection using the pcolor plot.


Regroupez les paramètres « GPU », « CPU_core » et « Price » pour créer un tableau croisé dynamique et visualiser cette connexion à l’aide du tracé pcolor.

````python
# Write your code below and press Shift+Enter to execute
# Create the group
df_gptest = df[[ "GPU", "CPU_core" ,"Price"]]
grouped_test1 = df_gptest.groupby([ "GPU", "CPU_core"], as_index=False).mean()
print(grouped_test1 )
````

<summary>Click here for Solution</summary>


</details>


````python
# Write your code below and press Shift+Enter to execute
# Create the Pivot table
grouped_pivot = grouped_test1.pivot(index="GPU", columns="CPU_core")
print(grouped_pivot)
````

````python
# Write your code below and press Shift+Enter to execute
# Create the Plot
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

fig.colorbar(im)
````

# Task 4 - Pearson Correlation and p-values

Use the `scipy.stats.pearsonr()` function to evaluate the Pearson Coefficient and the p-values for each parameter tested above. This will help you determine the parameters most likely to have a strong effect on the price of the laptops.


Utilisez la fonction scipy.stats.pearsonr() pour évaluer le coefficient de Pearson et les valeurs p pour chaque paramètre testé ci-dessus. Cela vous aidera à déterminer les paramètres les plus susceptibles d’avoir un effet fort sur le prix des ordinateurs portables.

````python
# Write your code below and press Shift+Enter to execute
for param in ['RAM_GB','CPU_frequency','Storage_GB_SSD','Screen_Size_inch','Weight_pounds','CPU_core','OS','GPU','Category']:
    pearson_coef, p_value = stats.pearsonr(df[param], df['Price'])
    print(param)
    print("The Pearson Correlation Coefficient for ",param," is", pearson_coef, " with a P-value of P =", p_value)
````
